```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

t = np.linspace(0,10,1000)
h0 = 100
v0 = 25
g = -9.8
v = v0 + g * t
h = h0 + v0 * t + g / 2 * t ** 2
h = np.clip(h, 0, 1000)
v[np.where(h==0)] = 0
```

```python
fig = plt.figure()
subplot_v = fig.add_subplot(2,1,1)
subplot_h = fig.add_subplot(2,1,2)
subplot_v.plot(t, v)
subplot_h.plot(t, h)
```

```python
# トラックの速度（km/h）
v = 200
# 単位の変換(m/s)
s = v * 1000 / 3600
# 重力加速度9.8で加速していって、sに達するまでの時間
t = s / 9.8
# 重力加速度9.8で加速していって、t秒後の移動距離
h = 4.9 * t ** 2
# ビル何階に相当するか（1回を3.5メートルで換算）
rank = h / 3.5 + 1
rank
```
